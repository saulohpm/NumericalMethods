📓 Cell 1 – Imports

```python
import numpy as np
from time import perf_counter
from scipy.integrate import quad

# Imported library already installed with pip.
import numerical_methods as nm
```

📓 Cell 2 – Problem setup

```python
# Integration Interal
a = -np.pi
b = np.pi
L = (b - a) / 2
C = (a + b) / 2

# Interval Subdivisions or probability
n = 12

# Domain
x = np.linspace(a, b, 1000)

# Step
delta_x = 0.001

# Function
f = lambda x: np.cos(x)
```

📓 Cell 3 – Available methods

```python
integration_methods = {
    "Rectangle Rule": nm.rectangle_integrate,
    "Trapezoid Rule": nm.trapezoidal_integrate,
    "First Simpson Rule": nm.simpson1_integrate,
    "Second Simpson Rule": nm.simpson2_integrate,
    "Gauss-Legendre Quadrature": nm.gauss_legendre_integrate,
    "Monte Carlo": nm.monte_carlo_integrate
}

differentiation_methods = {
    "Forward Difference": nm.fd.forward,
    "Backward Difference": nm.fd.backward,
    "Central Difference": nm.fd.central,
}
```

📓 Cell 4 – Numerical Integration

```python
print("#" * 60)
print(f"{'NUMERICAL INTEGRATION':^60}")
print("#" * 60)

for name, method in integration_methods.items():

    start = perf_counter()

    result = method(f, a, b, n)

    elapsed = perf_counter() - start

    print(f"{name:30} : {result:.6f}")
    print(f"{'Execution Time':30} : {elapsed*1000:.4f} ms\n")
```

📓 Cell 5 – Comparison with SciPy

```python
reference = quad(f, a, b)[0]

print("Reference (SciPy quad)")
print(f"{reference:.6f}")
```

📓 Cell 6 – Numerical Differentiation

```python
print("#" * 60)
print(f"{'NUMERICAL DIFFERENTIATION':^60}")
print("#" * 60)

evaluation_point = 2

for name, method in differentiation_methods.items():

    start = perf_counter()

    derivative = method(f, evaluation_point, delta_x)

    elapsed = perf_counter() - start

    print(f"{name:30} : {derivative:.6f}")
    print(f"{'Execution Time':30} : {elapsed * 1000:.4f} ms\n")
```

📓 Cell 8 – Series Approximation

```python
fourier_approximation = nm.fourier_approx(f, x, L, n)
taylor_aproximation = nm.taylor_approx(f, x, C, 7)
```

📓 Cell 8 – Graph

```python
nm.plot_function(f, x, a, b, fourier_approximation)
```
